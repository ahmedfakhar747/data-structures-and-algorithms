## ❓ Problem Statement

> Imagine you're building your ultimate football squad ⚽ with a limited budget.

Each player has a **cost** and a **rating**, and you want to pick the best combination such that:

* Total cost stays within your budget 💰
* Total team rating is maximized ⭐

There’s no restriction on team size — just choose wisely.


### 🧠 General Form

In computer science terms, this becomes the **Knapsack Problem**:

* You are given **n items**
* Each item has:

  * ⚖️ weight
  * 💎 profit

🎯 Goal:

* Total weight ≤ **W (capacity)**
* Maximize total profit

* Resource allocation

## 🧪 Testing Function

We can test our implementation in two simple ways:

* Directly call the function with inputs
* Use the `evaluate_test_case` helper for structured testing

💡 **Note:**
This is just test harness setup — feel free to skip it if you’re only focused on understanding the core logic.


````python
import time


def evaluate_test_case(function, test_case):
    try:
        start = time.perf_counter()
        result = function(**test_case["input"])
        end = time.perf_counter()

        expected = test_case["output"]
        passed = result == expected
        time_used = end - start

        return result, passed, time_used, None

    except Exception as e:
        return None, False, 0.0, e

def evaluate_test_cases(function, test_cases, error_only=False, summary_only=False):
    print("=" * 100)
    results = []
    total_time = 0.0

    for i, test_case in enumerate(test_cases):
        result, passed, time_used, error = evaluate_test_case(function, test_case)
        results.append((result, passed, time_used, error))
        total_time += time_used

        if summary_only:
            continue
        if error_only and passed:
            continue

        print(f"\n\033[1mTEST CASE #{i}\033[0m")
        print("Input:", test_case["input"])
        print("Expected:", test_case["output"])

        if error:
            print("\033[91mError:\033[0m", error)
        elif passed:
            print(f"\033[92mPassed ✔\033[0m  Time: {time_used:.6f}s")
        else:
            print(f"\033[91mFailed ✘\033[0m  Time: {time_used:.6f}s")
            print("Got:", result)

    total = len(results)
    passed_count = sum(1 for r in results if r[1])

    print("\n\033[1mSUMMARY\033[0m")
    print(
        f"TOTAL: {total}, \033[92mPASSED\033[0m: {passed_count}, \033[91mFAILED\033[0m: {total - passed_count}"
    )
    print(f"TOTAL TIME: {total_time:.2f}s")
    print("=" * 100)
    return results
````

## 🌀 Solution — Recursive Approach

<img src="https://i.imgur.com/eaJzv02.png" width="400">

We solve this problem by thinking **decision by decision** at each index.

At every step, we ask:

> “Should I take this item… or skip it?”


### 🔁 Core Idea

We define a recursive function:

```python
max_profit(idx, capacity)
```

This means:

> Maximum profit we can get starting from index `idx` with remaining capacity.


### ⚡ Decisions at Each Step

For every item, we have two choices:

* ❌ **Skip the item**
  Move to next index without changing capacity

* ✅ **Pick the item** *(only if it fits)*
  Add its profit and reduce remaining capacity


### 🔍 Cases

* If `weights[idx] > capacity`
  → We **cannot pick** it, so just skip

* Otherwise:
  → Take the **maximum** of:

  * Skipping the item
  * Picking the item


### 🛑 Base Case

* If we reach the end (`idx == n`)
  → No items left → profit = **0**


### 🧩 Recursion Tree (Intuition)

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20260218182009914083/group_3361.webp" width="700">

This tree shows how we explore **all possible combinations** — which is why this approach is simple but not efficient.

💡 This approach is intuitive, but it explores **all subsets (2ⁿ)** — we’ll optimize this next using DP.


###  💻 Implementation

````python
def max_profit_recursive(weights, profits, capacity, idx=0):
    if idx == len(weights):
        return 0
    elif weights[idx] > capacity:
        return max_profit_recursive(weights, profits, capacity, idx+1)
    else:
        option1 = max_profit_recursive(weights, profits, capacity, idx+1)
        option2 = profits[idx] + max_profit_recursive(weights, profits, capacity - weights[idx], idx+1)
        return max(option1, option2)
````

### 📝 Testing Our Algorithm:

````python
evaluate_test_cases(max_profit_recursive, tests)
````

### 📊 Analyzing the Complexity

Here’s what the recursion tree looks like for our approach:

<img src="https://i.imgur.com/WsKTC6I.png" width="640">

Each item creates **two branches** — one for **picking it** and one for **skipping it**.

💡 **Time Complexity:**

* In the worst case, we explore **all possible subsets**
* That’s **O(2ⁿ)** — exponential time

⚠️ For small inputs, recursion is fine, but it **gets slow very quickly** as n grows.
Next, we’ll see how **dynamic programming** can optimize this dramatically.

## 🧠 Solution — Memoization Approach

Recursion is simple, but it **recomputes the same subproblems** again and again.

💡 **Idea:** Store results of subproblems in a **memo table** so we can reuse them instead of recalculating.

* Each subproblem is uniquely defined by `(idx, capacity)`
* If we’ve already solved it, **return the stored result**
* Otherwise, compute it recursively and **save it for future use**

This converts our **exponential recursive solution** into a much faster **polynomial-time algorithm**.


````python
def max_profit_memo(weights, profits, capacity):
    memo = {}

    def recurse(idx, remaining):
        key = (idx, remaining)

        if key in memo:
            return memo[key]
            
        elif idx == len(weights):
            memo[key] = 0
            
        elif weights[idx] > remaining:
            memo[key] = recurse(idx+1, remaining)

        else:
            memo[key] = max(recurse(idx+1, remaining),
                           profits[idx] + recurse(idx+1, remaining-weights[idx]))
        return memo[key]

    return recurse(0, capacity)
````

### Testing Our Algorithm:

````python
evaluate_test_cases(max_profit_memo, tests)
````

## 📈 Solution — Dynamic Programming

Instead of recomputing subproblems, we build the solution **bottom-up** using a table.

### 🧠 Core Idea

We create a 2D table:

* Size → `(n + 1) × (capacity + 1)`
* `table[i][c]` = maximum profit using first `i` items with capacity `c`

Initially, everything is **0** (base case: no items or zero capacity).

### 🔁 How We Fill the Table

We iterate **row by row**, considering one item at a time.

For each cell `table[i][c]`:

### 🔍 Cases

* If the current item is **too heavy** (`weights[i-1] > c`)
  → ❌ Cannot pick it
  → Copy value from above:
  `table[i][c] = table[i-1][c]`

* Otherwise, we have two choices:

  * ❌ **Skip the item**
    → Profit = `table[i-1][c]`

  * ✅ **Pick the item**
    → Profit = `profits[i-1] + table[i-1][c - weights[i-1]]`

* Take the maximum of both:

---

### 🚀 Why This Works

* Avoids recomputation (unlike recursion)
* Builds solution step-by-step
* Time Complexity drops to **O(n × capacity)**


````python
def max_profit_dp(weights, profits, capacity):
    n = len(weights)
    table = [ [0 for _ in range(capacity+1) ] for _ in range(n+1)]
    for i in range(n):
        for c in range(1, capacity+1):
            if weights[i] > c:
                table[i+1][c] = table[i][c]
            else:
                table[i+1][c] = max(table[i][c],
                                   profits[i] + table[i][c-weights[i]])
    return table[-1][-1]
````

### Testing Our Algorithm:

````python
evaluate_test_cases(max_profit_dp, tests)
````

## 🎯 Conclusion

We explored the Knapsack Problem through three approaches — each leveling up in efficiency:

* **Recursion** 🐢
  Simple and intuitive, but explores all possibilities → **O(2ⁿ)** (slow)

* **Memoization** ⚡
  Avoids repeated work using caching → **much faster (O(n × capacity))**, but still uses recursion

* **Dynamic Programming** 🚀
  Fully optimized approach with **iterative solution** → fastest and most practical
