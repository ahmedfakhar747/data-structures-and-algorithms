## Testing Function:

We can test the function by passing the input to it directly or by using the `evaluate_test_case` function as written.

**Note:** This is test harness stuff. Ignore it if you’re just focusing on the logic.

```python
import time


def evaluate_test_case(function, test_case):
    try:
        start = time.perf_counter()
        result = function(**test_case["input"])
        end = time.perf_counter()

        expected = test_case["output"]
        passed = result == expected
        time_used = end - start

        return result, passed, time_used, None

    except Exception as e:
        return None, False, 0.0, e

def evaluate_test_cases(function, test_cases, error_only=False, summary_only=False):
    print("=" * 100)
    results = []
    total_time = 0.0

    for i, test_case in enumerate(test_cases):
        result, passed, time_used, error = evaluate_test_case(function, test_case)
        results.append((result, passed, time_used, error))
        total_time += time_used

        if summary_only:
            continue
        if error_only and passed:
            continue

        print(f"\n\033[1mTEST CASE #{i}\033[0m")
        print("Input:", test_case["input"])
        print("Expected:", test_case["output"])

        if error:
            print("\033[91mError:\033[0m", error)
        elif passed:
            print(f"\033[92mPassed ✔\033[0m  Time: {time_used:.6f}s")
        else:
            print(f"\033[91mFailed ✘\033[0m  Time: {time_used:.6f}s")
            print("Got:", result)

    total = len(results)
    passed_count = sum(1 for r in results if r[1])

    print("\n\033[1mSUMMARY\033[0m")
    print(
        f"TOTAL: {total}, \033[92mPASSED\033[0m: {passed_count}, \033[91mFAILED\033[0m: {total - passed_count}"
    )
    print(f"TOTAL TIME: {total_time:.2f}s")
    print("=" * 100)
    return results
```

## Solution - using recursive approach:

<img src="https://i.imgur.com/eaJzv02.png" width="400">

1. We'll write a recursive function that computes `max_profit(weights[idx:], profits[idx:], capacity)`, with `idx` starting from 0.


2. If `weights[idx] > capacity`, the current element is cannot be selected, so the maximum profit is the same as `max_profit(weights[idx+1:], profits[idx+1:], capacity)`.


3. Otherwise, there are two possibilities: we either pick `weights[idx]` or don't. We can recursively compute the maximum

    A. If we don't pick `weights[idx]`, once again the maximum profit for this case is `max_profit(weights[idx+1:], profits[idx+1:], capacity)`

    B. If we pick `weights[idx]`, the maximum profit for this case is `profits[idx] + max_profit(weights[idx+1:], profits[idx+1:], capacity - weights[idx]`


4. If `weights[idx:]` is empty, the maximum profit for this case is 0.

Here's a visual representation of recursive approach:

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20260218182009914083/group_3361.webp" width="700">

###  Implement the solution:

```python
def max_profit_recursive(weights, profits, capacity, idx=0):
    if idx == len(weights):
        return 0
    elif weights[idx] > capacity:
        return max_profit_recursive(weights, profits, capacity, idx+1)
    else:
        option1 = max_profit_recursive(weights, profits, capacity, idx+1)
        option2 = profits[idx] + max_profit_recursive(weights, profits, capacity - weights[idx], idx+1)
        return max(option1, option2)
```

### Testing our ALGORITHM:

```python
evaluate_test_cases(max_profit_recursive, tests)
```

## Solution - using memoization approach:

```python
def max_profit_memo(weights, profits, capacity):
    memo = {}

    def recurse(idx, remaining):
        key = (idx, remaining)

        if key in memo:
            return memo[key]
            
        elif idx == len(weights):
            memo[key] = 0
            
        elif weights[idx] > remaining:
            memo[key] = recurse(idx+1, remaining)

        else:
            memo[key] = max(recurse(idx+1, remaining),
                           profits[idx] + recurse(idx+1, remaining-weights[idx]))
        return memo[key]

    return recurse(0, capacity)
```

### Testing our ALGORITHM:

```python
evaluate_test_cases(max_profit_memo, tests)
```

## Solution - using dynamic programming:

1. Create a table of size `(n+1) * (capacity+1)` consisting of all 0s, where is `n` is the number of elements. `table[i][c]` represents the maximum profit that can be obtained using the first `i` elements if the maximum capacity is `c`. 

2. We'll fill the table row by row and column by column. `table[i][c]` can be filled using some values in the row above it.

3. If `weights[i] > c` i.e. if the current element can is larger than capacity, then `table[i][c]` is simply equal to `table[i-1][c]` (since there's no way we can pick this element).

4. If `weights[i] <= c` then we have two choices: to either pick the current element or not. We can compare the maximum profit for both these options and pick the better one as the value of `table[i][c]`.

    A. If we don't pick  the element with weight `weights[i]`, then once again the maximum profit is `table[i-1][c]`
    
    B. If we pick the element with weight `weights[i]`, then the maximum profit is `profits[i] + table[i-1][c-weights[i]]`, since we have used up some capacity.

```python
def max_profit_dp(weights, profits, capacity):
    n = len(weights)
    table = [ [0 for _ in range(capacity+1) ] for _ in range(n+1)]
    for i in range(n):
        for c in range(1, capacity+1):
            if weights[i] > c:
                table[i+1][c] = table[i][c]
            else:
                table[i+1][c] = max(table[i][c],
                                   profits[i] + table[i][c-weights[i]])
    return table[-1][-1]
```

### Testing our ALGORITHM:

```python
evaluate_test_cases(max_profit_dp, tests)
```

### Analysing the Complexity:

The complexity of the dynamic programming solution is **O(n × capacity)**

## Conclusion:

The Knapsack Problem was implemented using three different approaches: recursion, memoization, and dynamic programming. Each approach demonstrated clear differences in efficiency and scalability.

* **Recursive approach** is the most intuitive and easiest to implement, but it is highly inefficient due to repeated subproblem calculations, resulting in exponential time complexity.

* **Memoization (Top-Down DP)** significantly improves performance by caching previously computed results, reducing time complexity to polynomial. However, it still uses recursion, which adds overhead due to function calls and stack usage.

* **Dynamic Programming (Bottom-Up DP)** provides the most optimized solution. It eliminates recursion entirely and builds the solution iteratively, making it faster and more memory-efficient in practice.
