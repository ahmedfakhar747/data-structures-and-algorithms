## 🧪 Testing Function

We can test our implementation in two simple ways:

* Directly call the function with inputs
* Use the `evaluate_test_case` helper for structured testing

💡 **Note:**
This is just test harness setup — feel free to skip it if you’re only focused on understanding the core logic.


```python
import time


def evaluate_test_case(function, test_case):
    try:
        start = time.perf_counter()
        result = function(**test_case["input"])
        end = time.perf_counter()

        expected = test_case["output"]
        passed = result == expected
        time_used = end - start

        return result, passed, time_used, None

    except Exception as e:
        return None, False, 0.0, e

def evaluate_test_cases(function, test_cases, error_only=False, summary_only=False):
    print("=" * 100)
    results = []
    total_time = 0.0

    for i, test_case in enumerate(test_cases):
        result, passed, time_used, error = evaluate_test_case(function, test_case)
        results.append((result, passed, time_used, error))
        total_time += time_used

        if summary_only:
            continue
        if error_only and passed:
            continue

        print(f"\n\033[1mTEST CASE #{i}\033[0m")
        print("Input:", test_case["input"])
        print("Expected:", test_case["output"])

        if error:
            print("\033[91mError:\033[0m", error)
        elif passed:
            print(f"\033[92mPassed ✔\033[0m  Time: {time_used:.6f}s")
        else:
            print(f"\033[91mFailed ✘\033[0m  Time: {time_used:.6f}s")
            print("Got:", result)

    total = len(results)
    passed_count = sum(1 for r in results if r[1])

    print("\n\033[1mSUMMARY\033[0m")
    print(
        f"TOTAL: {total}, \033[92mPASSED\033[0m: {passed_count}, \033[91mFAILED\033[0m: {total - passed_count}"
    )
    print(f"TOTAL TIME: {total_time:.2f}s")
    print("=" * 100)
    return results
```

## 💥 Approach: Brute Force

### **How it works:**

- **Outer loop** picks every possible starting index `i`
- **Inner loop** extends the subarray from `i` to `j`, accumulating the sum
- As soon as `current_sum == target`, return the indices immediately
- If no subarray is found after all iterations, return `(None, None)`

### **Complexity:**

| | |
|---|---|
| Time | O(n³) — two nested loops + sum() recalculation inside |
| Space | O(1) — no extra data structures |

### Implementation:

```python
def subarray_sum_brute_force(array, target):
    n = len(array)
    
    for i in range(n):
        for j in range(i, n):
            if sum(array[i:j+1]) == target:
                return (i, j)
    
    return (None, None)
```

### Testing Our Function:

Let's walk through a quick example to see it in action.

```python
evaluate_test_cases(subarray_sum_brute_force, tests)
```

### Analyzing Inefficiencies:

The brute force approach runs in **O(n³)** time — the two nested loops contribute **O(n²)**, 
but calling `sum(array[i:j])` inside the inner loop adds another **O(n)** factor since it 
recalculates the entire subarray sum from scratch on every iteration.

## 🔧 Approach: Optimized Brute Force

### **How it works:**

- **Outer loop** picks every possible starting index `i`
- **Inner loop** extends the subarray from `i` to `j`, accumulating the sum
- As soon as `current_sum == target`, return the indices immediately
- If no subarray is found after all iterations, return `(None, None)`

### **Complexity:**

| | |
|---|---|
| Time | O(n²) — two nested loops |
| Space | O(1) — no extra data structures |


### Implementation

```python
def subarray_sum_optimized_brute_force(array, target):
    n = len(array)

    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += array[j]  # running sum → O(1) instead of O(n)
            if current_sum == target:
                return (i, j)
            if current_sum > target: # early termination
                break

    return (None, None)
```

### Testing Our Function:

Let's walk through a quick example to see it in action.

```python
evaluate_test_cases(subarray_sum_optimized_brute_force, tests)
```

### Analyzing Inefficiencies:

* Still **O(n²)** → two nested loops
* Early break only helps **average case**, not worst case
* Checks too many subarrays

## ⚡ Approach: Sliding window

### **How it works:**

### How it works (Sliding Window)

- **Start window**: Initialize `i = 0`, `j = 0`, `current_sum = 0`
- **Expand window**: Move `j` right and add `array[j]` to `current_sum`
- **Shrink window**: If `current_sum > target`, remove `array[i]` and move `i` right
- **Check condition**: If `current_sum == target`, return `(i, j-1)`

### **Complexity:**

| | |
|---|---|
| Time | O(n) — Each element processed at most twice |
| Space | O(1) — no extra data structures |


### Implementation:

```python
def subarray_sum_sliding_window(array, target):
    n = len(array)
    i, j, current_sum = 0, 0, 0

    while i < n and j < n+1:
        if current_sum == target:
            return (i, j-1)
        elif current_sum < target:
            if j < n:
                current_sum += array[j]
            j += 1
        elif current_sum > target:
            current_sum -= array[i]
            i += 1
    return (None, None)
```

```python
evaluate_test_cases(subarray_sum_sliding_window, tests)
```

---

## 🏁 Conclusion

The three approaches show a clear optimization progression:

* **Brute Force (O(n³))** → Recomputes sums repeatedly, leading to maximum redundancy and the worst performance.
* **Optimized Brute Force (O(n²))** → Uses running sum to reduce repeated calculations, but still explores all subarrays.
* **Sliding Window (O(n))** → Maintains a dynamic window and processes each element at most once, achieving optimal efficiency.

**Overall Insight:** Each step eliminates unnecessary computation, moving from exhaustive search → partial optimization → fully linear solution.
