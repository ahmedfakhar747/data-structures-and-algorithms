## 📌 Problem Statement

> Given an array of non-negative integers, find a continuous subarray that adds up to a given target sum.

![](https://i.imgur.com/tLEBF7x.png)

---

## 🧠 Intuition

* Brute force = try all subarrays → too slow (O(n²))
* Since all numbers are **non-negative**, we can use a **sliding window**
* Expand window → increase sum
* Shrink window → decrease sum


---

## 📥 Input

* **`array`** → A list of numbers (cannot contain non-negative integers)
* **`target`** → The required sum to be formed using a **contiguous subarray**

---

* **`end_index`** → Index where the valid subarray ends

## 🧪 Testing Function

We can test our implementation in two simple ways:

* Directly call the function with inputs
* Use the `evaluate_test_case` helper for structured testing

💡 **Note:**
This is just test harness setup — feel free to skip it if you’re only focused on understanding the core logic.


````python
import time


def evaluate_test_case(function, test_case):
    try:
        start = time.perf_counter()
        result = function(**test_case["input"])
        end = time.perf_counter()

        expected = test_case["output"]
        passed = result == expected
        time_used = end - start

        return result, passed, time_used, None

    except Exception as e:
        return None, False, 0.0, e

def evaluate_test_cases(function, test_cases, error_only=False, summary_only=False):
    print("=" * 100)
    results = []
    total_time = 0.0

    for i, test_case in enumerate(test_cases):
        result, passed, time_used, error = evaluate_test_case(function, test_case)
        results.append((result, passed, time_used, error))
        total_time += time_used

        if summary_only:
            continue
        if error_only and passed:
            continue

        print(f"\n\033[1mTEST CASE #{i}\033[0m")
        print("Input:", test_case["input"])
        print("Expected:", test_case["output"])

        if error:
            print("\033[91mError:\033[0m", error)
        elif passed:
            print(f"\033[92mPassed ✔\033[0m  Time: {time_used:.6f}s")
        else:
            print(f"\033[91mFailed ✘\033[0m  Time: {time_used:.6f}s")
            print("Got:", result)

    total = len(results)
    passed_count = sum(1 for r in results if r[1])

    print("\n\033[1mSUMMARY\033[0m")
    print(
        f"TOTAL: {total}, \033[92mPASSED\033[0m: {passed_count}, \033[91mFAILED\033[0m: {total - passed_count}"
    )
    print(f"TOTAL TIME: {total_time:.2f}s")
    print("=" * 100)
    return results
````

## ⚡ Approach: Sliding Window

* Maintain two pointers: `start` and `end`
* Keep track of current sum
* Move `end` to expand window
* Move `start` to shrink when sum exceeds target

---

## 🧩 Algorithm Steps

1. Initialize `start = 0`, `current_sum = 0`
2. Iterate `end` over array
3. Add `arr[end]` to `current_sum`
4. While `current_sum > target`, move `start` forward
5. If `current_sum == target`, return subarray

---

## 💻 Implementation

```python
def subarray_with_given_sum(arr, target):
    start = 0
    current_sum = 0

    for end in range(len(arr)):
        current_sum += arr[end]

        while current_sum > target and start <= end:
            current_sum -= arr[start]
            start += 1

        if current_sum == target:
            return arr[start:end+1]

    return None
```

---

## 🧪 Example

```python
arr = [1, 4, 20, 3, 10, 5]
target = 33

print(subarray_with_given_sum(arr, target))
```
