## 💻 Graph Representation

Before applying algorithms, we need a way to represent graphs in code.

The most common method are:
- **Edge List**

    In edge list we store graph as a list of node pairs, which is simple but inefficient for finding neighbors or traversing.

    Below is the implementation:

```python
# The nodes are labeled from `0` to `4`.
num_nodes = 5

'''
This list defines the connections between nodes.  
Each tuple `(u, v)` represents an edge between node `u` and node `v`.
  For example:
  - `(0, 1)` means node 0 is connected to node 1
  - `(1, 3)` means node 1 is connected to node 3

'''
edges = [(0, 1), (0, 4), (1, 2), (1, 3), (1, 4), (2, 3), (3, 4)]
num_edges = len(edges)

#This prints the number of nodes and edges as a tuple.
print(f"({num_nodes}, {num_edges})")
```

- **Adjacency List**

    <img src="https://i.imgur.com/rgMwkIW.png" width="700">

    In an adjacency list, each node stores a list of its neighbors.
  
    This approach is efficient in terms of space and is widely used in practice.

    Here is the implementation using class to represent a graph as an adjacency list in Python:

```python
class GraphAdjList:
    def __init__(self, num_nodes, edges):
        self.num_nodes = num_nodes
        self.data = [ [] for _ in range(num_nodes)]
        for n1, n2 in edges:
            self.data[n1].append(n2)
            self.data[n2].append(n1)

    def __repr__(self):
        return "\n".join([ f"{n}: {neibhbors}" for n, neibhbors in enumerate(self.data)])

    def __str__(self):
        return self.__repr__()
```

```python
graph1 = GraphAdjList(num_nodes, edges)
print(graph1)
```

> **Question**: Write a function to add an edge to a graph represented as an adjacency list. 

```python
def add_edge(graph, u, v):
    if v not in graph.data[u]:
        graph.data[u].append(v)
        graph.data[u].sort()
    if u not in graph.data[v]:
        graph.data[v].append(u)
        graph.data[v].sort()

add_edge(graph1, 0, 3)
print(graph1)
```

> **Question**: Write a function to remove an edge from a graph represented as a adjacency list.

```python
def remove_edge(graph, u, v):
    if v in graph.data[u]:
        graph.data[u].remove(v)
    if u in graph.data[v]:
        graph.data[v].remove(u)

remove_edge(graph1, 0, 3)
print(graph1)
```

- **Adjacency Matrix**

    <img src="https://i.imgur.com/oswYKTW.png" width="700">

    An adjacency matrix is a 2D grid where each cell shows if two nodes are connected, making edge checks fast but using more space.

    Below is a Python implementation of a graph using an adjacency matrix:

```python
class GraphAdjMatrix:
    def __init__(self, num_nodes, edges):
        self.num_nodes = num_nodes
        self.matrix = [ [ 0 for _ in range(num_nodes)] for _ in range(num_nodes)]
        for n1, n2 in edges:
            self.matrix[n1][n2] = 1
            self.matrix[n2][n1] = 1

    def __repr__(self):
        # Header row:
        header =  "    " + "  ".join(str(i) for i in range(num_nodes)) + "\n" + "    " + "  ".join("-" for i in range(num_nodes))

        rows = []
        for i in range(num_nodes):
            row = "  ".join(str(val) for val in self.matrix[i])
            rows.append(f"{i} | {row}")

        return header + "\n" + "\n".join(rows)

    def __str__(self):
        return self.__repr__()
```

```python
graph2 = GraphAdjMatrix(num_nodes, edges)
print(graph2)
```

> **Question**: Write a function to add an edge to a graph represented as an adjacency matrix.

```python
def add_edge(graph, u, v):
    if graph.matrix[u][v] == 0:
        graph.matrix[u][v] = 1
    if graph.matrix[v][u] == 0:
        graph.matrix[v][u] = 1

add_edge(graph2, 0, 3)
print(graph2)
```

> **Question**: Write a function to remove an edge to a graph represented as an adjacency matrix.

```python
def remove_edge(graph, u, v):
    if graph.matrix[u][v] == 1:
        graph.matrix[u][v] = 0
    if graph.matrix[v][u] == 1:
        graph.matrix[v][u] = 0

remove_edge(graph2, 0, 3)
print(graph2)
```

## 🔍 Graph Traversal

Graph traversal is the process of visiting all nodes in a graph systematically, typically using algorithms like **Breadth-First Search (BFS)** or **Depth-First Search (DFS)**.

## 🪜 Breadth First Search (BFS)

Breadth First Search (BFS) is a traversal algorithm that explores a graph level by level.

It starts from a source node and visits all its neighbors before moving to the next level.

<img src="https://he-s3.s3.amazonaws.com/media/uploads/fdec3c2.jpg" width=700>

### Key Characteristics:
- Uses a **queue** (FIFO)
- Guarantees the shortest path in unweighted graphs
- Explores nodes in layers

### Real-World Use Cases:
- Finding the shortest path
- Network broadcasting
- Web crawling

Here’s a clean Python example of BFS (Breadth-First Search) for a graph using an adjacency list:

![](https://i.imgur.com/xkgMnwx.png)

```python
def bfs(graph, start):
    # Stores nodes to visit
    queue = []
    #Prevents revisiting nodes
    discovered = [False] * len(graph.data)
    distance = [None] * len(graph.data)
    parent = [None] * len(graph.data)

    #Mark start as visited and push it into queue
    discovered[start] = True
    queue.append(start)
    distance[start] = 0
    idx = 0

    while idx < len(queue):
        # Move pointer forward(Dequeue)
        current = queue[idx]
        idx += 1

        # Loop over all adjacent nodes
        for node in graph.data[current]:
            # Only process unvisited nodes
            if not discovered[node]:
                # Distance increases level by level
                distance[node] = 1 + distance[current]
                parent[node] = current
                discovered[node] = True
                queue.append(node)

    return queue, distance, parent
```

```python
graph1 = GraphAdjList(num_nodes, edges)
print(bfs(graph1, 0))
```

## 🌀 Depth First Search (DFS)

Depth First Search (DFS) is a traversal algorithm that explores as far as possible along each branch before backtracking.

Instead of exploring level by level, DFS dives deep into the graph.

<img src="https://www.uvm.edu/~cbcafier/cs2240/content/assets/traversal2.jpg" width=700>

### Key Characteristics:
- Uses **recursion** or a stack
- Explores one path completely before switching
- Does not guarantee shortest path

### Real-World Use Cases:
- Path finding
- Cycle detection
- Solving puzzles (like mazes)

---

## 🟦 BFS vs DFS

| Feature            | BFS                     | DFS                     |
|------------------|------------------------|------------------------|
| Data Structure    | Queue                  | Stack / Recursion      |
| Traversal Style   | Level by level         | Deep first             |
| Shortest Path     | Yes (unweighted graph) | No                     |
| Memory Usage      | Higher                 | Lower                  |

Both algorithms are important and are used depending on the problem requirements.
