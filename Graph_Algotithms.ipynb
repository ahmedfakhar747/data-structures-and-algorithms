## 💻 Graph Representation

Before applying algorithms, we need a way to represent graphs in code.

The most common method are:
- **Edge List**

    In edge list we store graph as a list of node pairs, which is simple but inefficient for finding neighbors or traversing.

    Below is the implementation:

```python
# The nodes are labeled from `0` to `4`.
num_nodes = 5

'''
This list defines the connections between nodes.  
Each tuple `(u, v)` represents an edge between node `u` and node `v`.
  For example:
  - `(0, 1)` means node 0 is connected to node 1
  - `(1, 3)` means node 1 is connected to node 3

'''
edges = [(0, 1), (0, 4), (1, 2), (1, 3), (1, 4), (2, 3), (3, 4)]
num_edges = len(edges)

#This prints the number of nodes and edges as a tuple.
print(f"({num_nodes}, {num_edges})")
```

- **Adjacency List**

    <img src="https://i.imgur.com/rgMwkIW.png" width="700">

    In an adjacency list, each node stores a list of its neighbors.
  
    This approach is efficient in terms of space and is widely used in practice.

    Here is the implementation using class to represent a graph as an adjacency list in Python:

```python
class GraphAdjList:
    def __init__(self, num_nodes, edges):
        self.num_nodes = num_nodes
        self.data = [ [] for _ in range(num_nodes)]
        for n1, n2 in edges:
            self.data[n1].append(n2)
            self.data[n2].append(n1)

    def __repr__(self):
        return "\n".join([ f"{n}: {neibhbors}" for n, neibhbors in enumerate(self.data)])

    def __str__(self):
        return self.__repr__()
```

```python
graph1 = GraphAdjList(num_nodes, edges)
print(graph1)
```

> **Question**: Write a function to add an edge to a graph represented as an adjacency list. 

```python
def add_edge(graph, u, v):
    if v not in graph.data[u]:
        graph.data[u].append(v)
        graph.data[u].sort()
    if u not in graph.data[v]:
        graph.data[v].append(u)
        graph.data[v].sort()

add_edge(graph1, 0, 3)
print(graph1)
```

> **Question**: Write a function to remove an edge from a graph represented as a adjacency list.

```python
def remove_edge(graph, u, v):
    if v in graph.data[u]:
        graph.data[u].remove(v)
    if u in graph.data[v]:
        graph.data[v].remove(u)

remove_edge(graph1, 0, 3)
print(graph1)
```

- **Adjacency Matrix**

    <img src="https://i.imgur.com/oswYKTW.png" width="700">

    An adjacency matrix is a 2D grid where each cell shows if two nodes are connected, making edge checks fast but using more space.

    Below is a Python implementation of a graph using an adjacency matrix:

```python
class GraphAdjMatrix:
    def __init__(self, num_nodes, edges):
        self.num_nodes = num_nodes
        self.matrix = [ [ 0 for _ in range(num_nodes)] for _ in range(num_nodes)]
        for n1, n2 in edges:
            self.matrix[n1][n2] = 1
            self.matrix[n2][n1] = 1

    def __repr__(self):
        # Header row:
        header =  "    " + "  ".join(str(i) for i in range(num_nodes)) + "\n" + "    " + "  ".join("-" for i in range(num_nodes))

        rows = []
        for i in range(num_nodes):
            row = "  ".join(str(val) for val in self.matrix[i])
            rows.append(f"{i} | {row}")

        return header + "\n" + "\n".join(rows)

    def __str__(self):
        return self.__repr__()
```

```python
graph2 = GraphAdjMatrix(num_nodes, edges)
print(graph2)
```

## 🧪 Testing Function

We can test our implementation in two simple ways:

* Directly call the function with inputs
* Use the `evaluate_test_case` helper for structured testing

```python
import time


def evaluate_test_case(function, test_case):
    try:
        start = time.perf_counter()
        result = function(**test_case["input"])
        end = time.perf_counter()

        expected = test_case["output"]
        passed = result == expected
        time_used = end - start

        return result, passed, time_used, None

    except Exception as e:
        return None, False, 0.0, e

def evaluate_test_cases(function, test_cases, error_only=False, summary_only=False):
    print("=" * 100)
    results = []
    total_time = 0.0

    for i, test_case in enumerate(test_cases):
        result, passed, time_used, error = evaluate_test_case(function, test_case)
        results.append((result, passed, time_used, error))
        total_time += time_used

        if summary_only:
            continue
        if error_only and passed:
            continue

        print(f"\n\033[1mTEST CASE #{i}\033[0m")
        print("Input:", test_case["input"])
        print("Expected:", test_case["output"])
        print("Got:", result)

        if error:
            print("\033[91mError:\033[0m", error)
        elif passed:
            print(f"\033[92mPassed ✔\033[0m  Time: {time_used:.6f}s")
        else:
            print(f"\033[91mFailed ✘\033[0m  Time: {time_used:.6f}s")

    total = len(results)
    passed_count = sum(1 for r in results if r[1])

    print("\n\033[1mSUMMARY\033[0m")
    print(
        f"TOTAL: {total}, \033[92mPASSED\033[0m: {passed_count}, \033[91mFAILED\033[0m: {total - passed_count}"
    )
    print(f"TOTAL TIME: {total_time:.2f}s")
    print("=" * 100)
    return results
```

## 🔍 Graph Traversal

Graph traversal is the process of visiting all nodes in a graph systematically, typically using algorithms like **Breadth-First Search (BFS)** or **Depth-First Search (DFS)**.


---

## 🪜 Breadth First Search (BFS)

Breadth First Search (BFS) is a traversal algorithm that explores a graph level by level.

It starts from a source node and visits all its neighbors before moving to the next level.

<img src="https://he-s3.s3.amazonaws.com/media/uploads/fdec3c2.jpg" width=700>

### Key Characteristics:
- Uses a **queue** (FIFO)
- Guarantees the shortest path in unweighted graphs
- Explores nodes in layers

### Real-World Use Cases:
- Finding the shortest path
- Network broadcasting
- Web crawling

Here’s a clean Python example of BFS (Breadth-First Search) for a graph using an adjacency list:

![](https://i.imgur.com/xkgMnwx.png)

```python
def bfs(graph, start):
    # Stores nodes to visit
    queue = []
    #Prevents revisiting nodes
    discovered = [False] * len(graph.data)
    distance = [None] * len(graph.data)
    parent = [None] * len(graph.data)

    # Mark start as visited and push it into queue
    discovered[start] = True
    queue.append(start)
    distance[start] = 0
    idx = 0

    while idx < len(queue):
        # Move pointer forward(Dequeue)
        current = queue[idx]
        idx += 1

        # Loop over all adjacent nodes
        for node in graph.data[current]:
            # Only process unvisited nodes
            if not discovered[node]:
                # Distance increases level by level
                distance[node] = 1 + distance[current]
                parent[node] = current
                discovered[node] = True
                queue.append(node)

    return queue, distance, parent
```

> **Question**: Write a function to detect a cycle in a graph

<img src="https://upload.wikimedia.org/wikipedia/commons/b/b5/Cycle_graph_C5.png" width="500">

```python
def has_cycle(graph):
    discovered = [False] * len(graph.data)
    parent = [None] * len(graph.data)

    for node in range(len(graph.data)):
        if not discovered[node]:
            queue = [node]       
            discovered[node] = True
            idx = 0
        
            while idx < len(queue):
                current = queue[idx]
                idx += 1
        
                for neighbour in graph.data[current]:
                    if not discovered[neighbour]:
                        parent[neighbour] = current
                        discovered[neighbour] = True
                        queue.append(neighbour)
                    elif neighbour != parent[current]:
                        return True
    return False
```

```python
# A square loop: 0 -> 1 -> 2 -> 3 -> 0
graph3 = {
    (0, 1),
    (1, 2),
    (2, 3),
    (3, 0)
}

# A simple path: 0 -> 1 -> 2 -> 3
graph4 = {
    (0, 1),
    (1, 2),
    (2, 3)
}

print(f"Is graph3 cyclic? {has_cycle(GraphAdjList(4, graph3))}")   # Output: True
print(f"Is graph4 cyclic? {has_cycle(GraphAdjList(4, graph4))}") # Output: False
```

## 🌀 Depth First Search (DFS)

Depth First Search (DFS) is a traversal algorithm that explores as far as possible along each branch before backtracking.

Instead of exploring level by level, DFS dives deep into the graph.

<img src="https://www.uvm.edu/~cbcafier/cs2240/content/assets/traversal2.jpg" width=700>

### Key Characteristics:
- Uses **recursion** or a stack
- Explores one path completely before switching
- Does not guarantee shortest path

### Real-World Use Cases:
- Path finding
- Cycle detection
- Solving puzzles (like mazes)

Here’s a clean Python example of BFS (Breadth-First Search) for a graph using an adjacency list:

```python
def dfs(graph, start):
    stack = [(start, 0, None)]  # (node, depth, parent)

    discovered = [False] * len(graph.data)

    order = []
    distance = [None] * len(graph.data)
    parent = [None] * len(graph.data)

    while stack:
        node, depth, par = stack.pop()

        if not discovered[node]:
            discovered[node] = True

            order.append(node)
            distance[node] = depth
            parent[node] = par

            for neighbor in reversed(graph.data[node]):
                if not discovered[neighbor]:
                    stack.append((neighbor, depth + 1, node))

    return (order, distance, parent)
```

A **directed graph** (aka **digraph**) is a graph where **edges have a direction**.

* If there’s an edge from `A → B`, you can go from A to B, **but not necessarily B → A**.

Formally:

* Graph = ( G = (V, E) )

### **Key differences from undirected graphs**

| Feature | Undirected   | Directed                                     |
| ------- | ------------ | -------------------------------------------- |
| Edge    | {u, v}       | (u, v)                                       |
| Travel  | Both ways    | Only the specified direction                 |

| Cycle   | Normal cycle | Directed cycle (all arrows follow direction) |

* **in-degree** = number of edges coming **into** a node
* **out-degree** = number of edges going **out** of a node

Some algorithms work differently or only make sense in digraphs:

* **Topological Sort** → only for DAGs (Directed Acyclic Graphs)

* **Strongly Connected Components (SCC)** → Kosaraju’s or Tarjan’s

## 🛠️ Flexible Graph Class

> **Question**: Write a Python class `Graph` that supports directed/undirected and weighted/unweighted graphs, stores edges in adjacency lists, and prints the graph neatly.

```python
class Graph():
    def __init__(self, num_nodes, edges, directed=False, weighted=False):
        self.num_nodes = num_nodes          # Total number of nodes
        self.directed = directed            # True → directed graph
        self.weighted = weighted            # True → weighted graph
        
        # Adjacency list: stores neighbors
        self.data = [[] for _ in range(num_nodes)]
        
        # Stores weights corresponding to edges (only used if weighted=True)
        self.weight = [[] for _ in range(num_nodes)]

        # Build the graph
        for edge in edges:
            if self.weighted:
                node1, node2, weight = edge   # Edge with weight
                
                self.data[node1].append(node2)
                self.weight[node1].append(weight)
                
                if not directed:              # Add reverse edge if undirected
                    self.data[node2].append(node1)
                    self.weight[node2].append(weight)
            else:
                node1, node2 = edge           # Edge without weight
                
                self.data[node1].append(node2)
                
                if not directed:              # Add reverse edge if undirected
                    self.data[node2].append(node1)

    def __repr__(self):
        result = ""
        if self.weighted:
            for i, (nodes, weights) in enumerate(zip(self.data, self.weight)):
                result += f"{i}: {list(zip(nodes, weights))}\n"
        else:
            for i, nodes in enumerate(self.data):
                result += f"{i}: {nodes}\n"
        return result

    def __str__(self):
        return repr(self)
```

## ⚡ Dijkstra’s Algorithm

<img src="https://www.norsemathology.org/mediawiki-1.38.1/images/8/8e/Ellis2.gif" width="650">

### Steps

1. **Initialize**

   * Mark all nodes as **unvisited**
   * Set distances:

     * Start node → `0`
     * Others → `∞`

2. **Select Current Node**

   * Pick the unvisited node with the **smallest distance**

3. **Relax Neighbors**

   * For each unvisited neighbor:

     * Compute:

       `
       new_distance = current_distance + edge_weight
       `
     * Update if the new distance is smaller

4. **Mark Visited**

   * Mark current node as **visited**
   * It won’t be checked again

5. **Check Stop Condition**
   Stop if:

   * Destination is reached
   * Remaining nodes have distance = `∞` (disconnected)

6. **Repeat**

   * Choose next closest unvisited node
   * Continue from step 3

### Rule

* Works only with **non-negative weights**

```python
def shortest_path(graph, start, target):
    """Find the length of the shortest path between source and destination"""
    visited = [False] * len(graph.data)
    parent = [None] * len(graph.data)
    distance = [float('inf')] * len(graph.data)
    queue = []

    distance[start] = 0
    queue.append(start)
    idx = 0

    while idx < len(queue) and not visited[target]:
        current = queue[idx]
        idx += 1

        # Update the distances of all the neighors
        update_distances(graph, current, distance, parent)
        # Find the first unvisited node with smallest distance
        next_node = pick_next_node(distance, visited)
        if next_node is not None:
            visited[next_node] = True
            queue.append(next_node)

    return distance[target], distance, parent


def update_distances(graph, start, distance, parent=None):
    """Update the distances of the current node's neighbors"""
    neighbors = graph.data[start]
    weights = graph.weight[start]
    for i, node in enumerate(neighbors):
        weight = weights[i]
        if distance[start] + weight < distance[node]:
            distance[node] = distance[start] + weight
            if parent:
                parent[node] = start


def pick_next_node(distance, visited):
    """Pick the next univisited node at the smallest distance"""
    min_distance = float('inf')
    min_node = None
    for node in range(len(distance)):
        if not visited[node] and distance[node] < min_distance:
            min_node = node
            min_distance = distance[node]
    return min_node
```

### Complexity

- `pick_next_node` → **O(V)** linear scan, called **V times** → **O(V²)**
- `update_distances` → **O(E)** total across all iterations
- Outer while loop → **O(V)** once per node
- **Total → O(V²)** dominated by `pick_next_node`

### Inefficiency

- **O(V²)** too slow for large graphs
- Scans **all nodes** every iteration even if already visited
- No priority ordering, brute forces the minimum every time

That's why there is another algorithm named **Binary Heap** as follows

## 🌳 Binary Heap

A data structure to maintain the running minimum/maximum of a set of numbers, supporting efficient addition/removal.


<img src="https://i.imgur.com/ABAcM7m.png" width="600">


Heap operations:

- Insertion - $O(log N)$
- Min/Max - $O(1)$ (depending on type of heap)
- Deletion - $O(log N)$
- Convert a list to a heap - $O(n)$


Python's built-in heap: https://docs.python.org/3/library/heapq.html

> **Question**: Implement Dijkstra's shortest path algorithm using the `heap` module from Python. What is the complexity of the algorithm?

```python
import heapq

def shortest_path_heap(graph, start, target):
    distance = [float('inf')] * len(graph.data)
    parent = [None] * len(graph.data)
    distance[start] = 0
    heap = [(0, start)]  # (distance, node)

    while heap:
        dist, current = heapq.heappop(heap)
        if current == target:
            break
        for i, neighbor in enumerate(graph.data[current]):
            weight = graph.weight[current][i]
            new_dist = dist + weight
            if new_dist < distance[neighbor]:
                distance[neighbor] = new_dist
                parent[neighbor] = current
                heapq.heappush(heap, (new_dist, neighbor))

    return distance[target], distance, parent
```

## ⚔️ Simple Dijkstra vs Heap Dijkstra

| | **Simple Dijkstra** | **Heap Dijkstra** |
|---|---|---|
| **Finding min node** | Linear scan O(V) | Heappop O(log V) |
| **Overall complexity** | O(V²) | O((V+E) log V) |
| **Small graphs** | Fine | Slight overhead |
| **Large sparse graphs** | Very slow | Much faster |
| **Code complexity** | Simple, easy to understand | Slightly more complex |
| **Memory** | Low | Extra heap storage |
